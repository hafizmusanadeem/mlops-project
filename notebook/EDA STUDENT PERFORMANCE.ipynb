8. Adding Columns for "Total Score" and "Average"

```python
data['total score'] = data['math_score'] + data['reading_score'] + data['writing_score'] 

data['average'] = data['total score'] / 3
data.head()
```

```python
reading_full = data[data['reading_score'] == 100]['average'].count()
writing_full = data[data['writing_score'] == 100]['average'].count()
math_full = data[data['math_score'] == 100]['average'].count()

print(f'Number of Students with full marks in Maths: {math_full}')
print(f'Number of Students with full marks in Writing: {writing_full}')
print(f'Number of Students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = data[data['reading_score'] <= 20]['average'].count()
writing_less_20 = data[data['writing_score'] <= 20]['average'].count()
math_less_20 = data[data['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

```python
# Group Data
grouped = data.groupby('race_ethnicity').mean(numeric_only=True)

# Prepare Figure
fig, axs = plt.subplots(1, 3, figsize = (20, 8))
fig.suptitle("Average Scores by Race/Ethnicity", fontsize = 24, fontweight = 'bold')

# Column names and styling configs
score_info = [
    ('math_score', "Math Score", "#000000"),
    ('reading_score', "Reading Score", "#000000"),
    ('writing_score', "Writing Score", "#000000")
]

# Loop through each score type
for ax, (col, title, color) in zip(axs, score_info):
    sns.barplot(
        x = grouped.index,
        y = grouped[col],
        palette = 'winter',
        ax = ax
    )

    ax.set_title(title, color = color, size = 18)
    ax.set_ylabel("Average Score")
    ax.set_xlabel("Race & Ethnicity")
    ax.tick_params(axis = 'x', rotation = 30)

    # Add bar labels
    for container in ax.containers:
        ax.bar_label(container, color = 'black', size = 12)

# Tight Layout
plt.tight_layout()
plt.show()
```

**Insights:**
* Group E students have scored the highest marks.
* Group A students have scored the lowest marks.
* Students from a lower Socioeconomic status have a lower avg in all course subjects.
